```python
import pandas as pd
from tools import load_data
import plotly.graph_objects as go
import plotly.express as px
pd.set_option("display.max_rows", None, "display.max_columns", None)
df = load_data()
df.info()
```

```python
#df.to_csv('out.csv')
common_property_types = df.property_type.value_counts()[:6].index.tolist()
df.property_type.value_counts()
df.property_type = df.property_type.apply(lambda x: x if x in common_property_types else "Other")
df.property_type.value_counts()
```

```python
df.to_csv('out.csv')
```

```python
common_property_types = df.property_type.value_counts()[:10].index.tolist()
property_type_price = pd.DataFrame(df[df.property_type.isin(common_property_types)].groupby('property_type')['price'].mean()).sort_values('price').reset_index()
fig = px.bar(property_type_price, x="price", y="property_type", title="Mean price by property type", orientation="h",
             labels={"price": "Mean price",
                     "property_type": "Property type"})
fig.update_layout(height=800)
fig.show()
```

```python
beds_price = pd.DataFrame(df.groupby('beds')['price'].mean()).reset_index()
fig = px.bar(beds_price, x="beds", y="price", title="Mean price by number of beds",
             labels={"price": "Mean price",
                     "beds": "Number of beds"})
fig.show()
```

```python
bedrooms_price = pd.DataFrame(df .groupby('bedrooms')['price'].mean()).reset_index()
fig = px.bar(bedrooms_price, x="bedrooms", y="price", title="Mean price by number of bedrooms",
             labels={"price": "Mean price",
                     "bedrooms": "Number of bedrooms"})
fig.show()
```

```python
bathrooms_price = pd.DataFrame(df.groupby('bathrooms')['price'].mean()).reset_index()
fig = px.bar(bathrooms_price, x="bathrooms", y="price", title="Mean price by number of bathrooms",
             labels={"price": "Mean price",
                     "bathrooms": "Number of bathrooms"})
fig.show()
```

```python
neighbourhood_price = pd.DataFrame(df.groupby('neighbourhood_cleansed')['price'].mean()).sort_values('price').reset_index()
fig = px.bar(neighbourhood_price, x="price", y="neighbourhood_cleansed", title="Mean price by neighbourhood", orientation="h",
             labels={"price": "Mean price per day",
                     "neighbourhood_cleansed": "Neighbourhood"})
fig.update_layout(height=800)
fig.show()
```

```python
room_type_price = pd.DataFrame(df.groupby('room_type')['price'].mean()).sort_values('price').reset_index()
fig = px.bar(room_type_price, x="price", y="room_type", title="Mean price by room type",
             labels={"price": "Mean price per day",
                     "room_type": "Room type"})
fig.show()
```

```python
room_type = df.groupby('room_type').count().reset_index().rename({"host_since": "number_of_listings"} ,axis = 1)[["room_type", "number_of_listings"]]
fig = px.bar(room_type, x="room_type", y="number_of_listings", title="Number of room type listed",
             labels={"room type": "type of room listed",
                     "value": "Number of listings"})
fig.show()
```
